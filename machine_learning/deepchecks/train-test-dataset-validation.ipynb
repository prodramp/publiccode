```python
from deepchecks.suites import full_suite
import pandas as pd
```

```python
source = 'https://raw.githubusercontent.com/prodramp/publiccode/master/datasets/titanic.csv'
#source = 'https://raw.githubusercontent.com/prodramp/publiccode/master/datasets/heart.csv'
```

```python
df = pd.read_csv(source)
```

```python
df
```

```python
df.describe()
```

```python
df.dtypes
```

```python
from sklearn.model_selection import train_test_split
```

```python
df_train, df_test = train_test_split(df, test_size=.005)
```

```python
print(df_train.shape)
print(df_test.shape)
```

```python
label_col = 'Survived'
```

```python
from deepchecks import Dataset
```

```python
ds_train = Dataset(df_train, label = label_col)
```

```python
ds_test = Dataset(df_test, label = label_col, cat_features= ['Pclass', 'Sex', 'Embarked', 'Parch', 'SibSp'])
```

```python
ds_train
```

```python
suite = full_suite()
```

```python
suite_result = suite.run(train_dataset=ds_train, test_dataset=ds_test)
```

```python
suite_result
```

```python
suite_result.save_as_html('titanic_full_test_bad.html')
```
