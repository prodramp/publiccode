```python
from deepchecks.suites import full_suite
import pandas as pd
```

```python
#source = 'https://raw.githubusercontent.com/prodramp/publiccode/master/datasets/titanic.csv'
source = 'https://raw.githubusercontent.com/prodramp/publiccode/master/datasets/heart.csv'
```

```python
df = pd.read_csv(source)
```

```python
df
```

```python
suite = full_suite()
```

```python
suite.run(df)
```

```python
sample_20 = df.sample(20)
```

```python
ndf = pd.concat([df, sample_20])
```

```python
ndf.reset_index()
```

```python
ndf
```

```python
suite.run(ndf)
```

```python
from deepchecks.suites import single_dataset_integrity
```

```python
intg_suite = single_dataset_integrity()
```

```python
suite_result = intg_suite.run(ndf)
```
